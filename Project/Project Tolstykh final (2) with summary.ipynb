```python
# Не учитывались категориальные столбцы, применим для них OneHotEncoding 

attriction_one_hot = pd.get_dummies(df_hr['Attrition'], prefix = 'attr', drop_first = True)
bussiness_travel_one_hot = pd.get_dummies(df_hr['BusinessTravel'], prefix = 'bus_tr')
gender_one_hot = pd.get_dummies(df_hr['Gender'], prefix = 'gend', drop_first = True)
marital_status_one_hot = pd.get_dummies(df_hr['MaritalStatus'], prefix = 'marital')
RelationshipSatisfaction_one_hot = pd.get_dummies(df_hr['RelationshipSatisfaction'], prefix = 'rel_sat')
Education_one_hot = pd.get_dummies(df_hr['Education'], prefix = 'edu')
EducationField_one_hot = pd.get_dummies(df_hr['EducationField'], prefix = 'edu_f')
Department_one_hot = pd.get_dummies(df_hr['Department'], prefix = 'dep')
JobRole_one_hot = pd.get_dummies(df_hr['JobRole'], prefix = 'job_role')

df_hr = df_hr.join([attriction_one_hot, 
            bussiness_travel_one_hot, 
            gender_one_hot, 
            marital_status_one_hot, 
            RelationshipSatisfaction_one_hot,
            Education_one_hot,
            EducationField_one_hot,
            Department_one_hot,
            JobRole_one_hot])
```

```python
#Самые важные признаки, влияющие на месячный доход
most_important_f = ['Age', 
 'JobLevel', 
 'NumCompaniesWorked', 
 'TotalWorkingYears', 
 'attr_Yes', 
 'job_role_Laboratory Technician', 
 'job_role_Manager',
 'job_role_Research Director',
 'job_role_Research Scientist',
 'job_role_Sales Representative']
```

# Подготовка данных для обучения

```python
most_important_f.extend(['job_role_Sales Executive', 
                                        'job_role_Human Resources',
                                        'job_role_Healthcare Representative',
                                        'bus_tr_Non-Travel',
                                        'bus_tr_Travel_Frequently',
                                        'bus_tr_Travel_Rarely',
                                        'gend_Male',
                                        'marital_Divorced',
                                        'marital_Married',
                                        'marital_Single',
                                        'rel_sat_1',
                                        'rel_sat_2',
                                        'rel_sat_3',
                                        'rel_sat_4',
                                        'edu_1',
                                        'edu_2',
                                        'edu_3',
                                        'edu_4',
                                        'edu_5',
                                        'edu_f_Human Resources',
                                        'edu_f_Life Sciences',
                                        'edu_f_Marketing',
                                        'edu_f_Medical',
                                        'edu_f_Other',
                                        'edu_f_Technical Degree'
                                       ])
```

```python
df_hr.shape
```

```python
X_train = df_hr.loc[:1200, most_important_f]
Y_train = df_hr.loc[:1200, 'MonthlyIncome']
```

```python
X_test = df_hr.loc[1200:, most_important_f]
Y_test = df_hr.loc[1200:, 'MonthlyIncome']
```

# Обучение

Построение модели линейной регресси

```python
lin_mod = LinearRegression()
```

```python
lin_mod.fit(X_train, Y_train)
```

```python
for pred_, true_ in zip(lin_mod.predict(X_test), Y_test):
    print(pred_, true_)
```

Построение модели опорных векторов

```python
svr_reg = SVR()
```

```python
svr_reg.fit(X_train, Y_train)
```

```python
for pred_, true_ in zip(svr_reg.predict(X_test), Y_test):
    print(pred_, true_)
```

```python
Регрессия с усилением градиента
```

```python
gb_reg = GradientBoostingRegressor()
```

```python
gb_reg.fit(X_train, Y_train)
```

```python
for pred_, true_ in zip(gb_reg.predict(X_test), Y_test):
    print(pred_, true_)
```

Посчитаем среднее квадратичное ошибку

```python
# Для метода опорных векторов
res = 0
for pred_, true_ in zip(svr_reg.predict(X_test), Y_test):
    res += (pred_ - true_) ** 2
    
res /= Y_test.shape[0]

res
```

```python
# Для линейной регрессии
res = 0
for pred_, true_ in zip(lin_mod.predict(X_test), Y_test):
    res += (pred_ - true_) ** 2
    
res /= Y_test.shape[0]

res
```

```python
# Для GradientBoostingRegressor()
res = 0
for pred_, true_ in zip(gb_reg.predict(X_test), Y_test):
    res += (pred_ - true_) ** 2
    
res /= Y_test.shape[0]

res
```
