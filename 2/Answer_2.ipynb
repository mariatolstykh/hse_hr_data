```python
# Эффективно выделим кусок данных через numpy where

df_hr['anomally'] = np.where(
                            df_hr['YearsAtCompany'] > 
                            np.mean(df_hr.YearsAtCompany) + 3*np.std(df_hr.YearsAtCompany),                 
                            1, 0)
```

```python
df_anomally = df_hr[df_hr['anomally'] == 1]

df_anomally
```

```python
print("Count of Anomally YearsAtWork is {} ".format(len(df_anomally)))
```

## Оценка почасовой ставки

```python
plt.hist(df_hr['HourlyRate'], color = 'pink', edgecolor = 'black', bins = 300)
```

Наблюдаются отклонения в значениях почасовой ставки

```python
df_hr.HourlyRate.quantile(q=0.9)
```

Это означает, что 90% сотрудников в выборке почасовая ставка менее 94 долларов.

Давайте объявим аномалиями все что выше этой отметки и выведем тех у кого 94 и более отсутствий. 

```python
df_hr_ab = df_hr[df_hr['HourlyRate'] > 93]
df_hr_ab
```

Получили 160 человек.
